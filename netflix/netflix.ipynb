```python
netflix.dropna(inplace = True)
netflix.isna().sum()
```

```python
# 필요한 변수를 찾는다. 
netflix_need = netflix[['type', 'title','country', 'release_year','rating','listed_in', 'description']]
netflix_need.head()
```

```python
netflix_need['type'].value_counts() # Movie or TV Show
```

```python
netflix_movie = netflix_need[netflix_need['type'] == 'Movie']
print(netflix_movie.shape)

netflix_movie.head() # 4673 by 7
```

```python
netflix_TVshow = netflix_need[netflix_need['type'] == 'TV Show']
print(netflix_TVshow.shape)

netflix_TVshow.head() # 135 by 7
```

```python
netflix_movie.head()
```

```python
netflix_movie_new = netflix_movie[netflix_movie['release_year'] > 2005]
netflix_movie_new.reset_index(inplace = True)
```

```python
del netflix_movie_new['index']
```

```python
netflix_movie_new.head().info()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(min_df = 2, max_df = 0.7)
tfidf_vec = tfidf.fit_transform(netflix_movie_new['description'])
print(tfidf.get_feature_names()[:7])
print(tfidf_vec.toarray())
```

```python
kim = [word for word in tfidf.get_feature_names() if word.isalpha() == True]
kim[:5]
```

```python
NEW_MOVIE = pd.DataFrame(tfidf_vec.toarray(), columns = tfidf.get_feature_names(), index = netflix_movie_new['title'])
NEW_MOVIE.head()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
net_movie = pd.DataFrame(cosine_similarity(tfidf_vec), columns = netflix_movie_new['title'], index = netflix_movie_new['title'])
net_movie.head()
```

```python
netflix_movie_new[netflix_movie_new['title']== 'About Time']
```

```python
def movie_recsys(movie, number):
    MOVIE_RECSYS = net_movie[movie].sort_values(ascending = False)
    return MOVIE_RECSYS[1:].head(number)
movie_recsys(movie = 'About Time', number = 10)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidfvec = TfidfVectorizer(min_df = 2, max_df = 0.7)
tvshow_vector = tfidfvec.fit_transform(netflix_TVshow['description'])
```

```python
# feature tokenizer 
tfidfvec.get_feature_names()[:5]
```

```python
print(tvshow_vector.toarray().shape)
print(tvshow_vector.toarray())
```

```python
# 각 영화별 단어가 얼마만큼 들어가있는지 맵핑해준다.
tvshow_vector = pd.DataFrame(tvshow_vector.toarray(), columns = tfidfvec.get_feature_names(), index = netflix_TVshow['title'])
tvshow_vector.head()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
TVshow = cosine_similarity(tvshow_vector)
TVshow_df = pd.DataFrame(TVshow, index = netflix_TVshow['title'], columns = netflix_TVshow['title'])
TVshow_df.head()
```
